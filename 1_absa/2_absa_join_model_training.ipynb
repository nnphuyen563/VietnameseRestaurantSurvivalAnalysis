```python
# Dataset Class
class ReviewDataset(Dataset):
    def __init__(self, data, tokenizer, max_length):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.iloc[idx]["clean_cmt"]
        food = self.data.iloc[idx]["a_food"]
        service = self.data.iloc[idx]["a_service"]
        atmosphere = self.data.iloc[idx]["a_atmosphere"]

        encoded = self.tokenizer(
            text,
            max_length=self.max_length,
            truncation=True,
            padding="max_length",
            return_tensors="pt"
        )

        return {
            "input_ids": encoded["input_ids"].squeeze(0),
            "attention_mask": encoded["attention_mask"].squeeze(0),
            "food": torch.tensor(food, dtype=torch.long),
            "service": torch.tensor(service, dtype=torch.long),
            "atmosphere": torch.tensor(atmosphere, dtype=torch.long)
        }
```

# Model

```python
from transformers import AutoModel, AutoTokenizer, AutoConfig
import torch.nn as nn
import torch

class AspectBasedSentimentModel(nn.Module):
    def __init__(self, model_name, pretrained = True):
        super(AspectBasedSentimentModel, self).__init__()

        if pretrained:
            self.encoder = AutoModel.from_pretrained(model_name)  # Pretrained transformer
        else:
            self.encoder = AutoModel.from_config(AutoConfig.from_pretrained(model_name))  # Randomly initialized transformer

        hidden_size = self.encoder.config.hidden_size

        self.food_head = nn.Linear(hidden_size, 5)  # Classification head for a_food
        self.service_head = nn.Linear(hidden_size, 5)  # Classification head for a_service
        self.atmosphere_head = nn.Linear(hidden_size, 5)  # Classification head for a_atmosphere

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]  # CLS token representation

        food_logits = self.food_head(cls_output)  # Logits for a_food
        service_logits = self.service_head(cls_output)  # Logits for a_service
        atmosphere_logits = self.atmosphere_head(cls_output)  # Logits for a_atmosphere

        return food_logits, service_logits, atmosphere_logits

# Tokenizer and data preparation
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AspectBasedSentimentModel(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

# Train

```python
from tqdm.auto import tqdm
from sklearn.metrics import accuracy_score, f1_score

# Training Loop with tqdm and aspect-wise MSE
def train_epoch(epoch, epochs, model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0

    progress_bar = tqdm(data_loader, desc=f"Epoch {epoch+1}/{epochs}", unit = "Batch", leave=False)

    for batch in progress_bar:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        # Adjust to 0-based indexing
        food = batch["food"].to(device) - 1
        service = batch["service"].to(device) - 1
        atmosphere = batch["atmosphere"].to(device) - 1

        optimizer.zero_grad()
        food_logits, service_logits, atmosphere_logits = model(input_ids, attention_mask)

        loss = (
            criterion(food_logits, food) +
            criterion(service_logits, service) +
            criterion(atmosphere_logits, atmosphere)
        )

        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        # Update progress bar with the current loss
        progress_bar.set_postfix(loss=loss.item())

    return total_loss / len(data_loader)

def evaluate_epoch(model, data_loader, criterion, device):
    model.eval()
    total_loss = 0
    food_preds, service_preds, atmosphere_preds = [], [], []
    food_true, service_true, atmosphere_true = [], [], []

    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            food = batch["food"].to(device) - 1  # Adjust to 0-based indexing
            service = batch["service"].to(device) - 1
            atmosphere = batch["atmosphere"].to(device) - 1

            food_logits, service_logits, atmosphere_logits = model(input_ids, attention_mask)

            loss = (
                criterion(food_logits, food) +
                criterion(service_logits, service) +
                criterion(atmosphere_logits, atmosphere)
            )
            total_loss += loss.item()

            food_preds.extend(torch.argmax(food_logits, dim=1).cpu().numpy())
            service_preds.extend(torch.argmax(service_logits, dim=1).cpu().numpy())
            atmosphere_preds.extend(torch.argmax(atmosphere_logits, dim=1).cpu().numpy())

            food_true.extend(food.cpu().numpy())
            service_true.extend(service.cpu().numpy())
            atmosphere_true.extend(atmosphere.cpu().numpy())

    # Accuracy for each aspect
    food_acc = accuracy_score(food_true, food_preds)
    service_acc = accuracy_score(service_true, service_preds)
    atmosphere_acc = accuracy_score(atmosphere_true, atmosphere_preds)

    # F1 score for each aspect
    food_f1 = f1_score(food_true, food_preds, average="weighted")
    service_f1 = f1_score(service_true, service_preds, average="weighted")
    atmosphere_f1 = f1_score(atmosphere_true, atmosphere_preds, average="weighted")

    overall_acc = accuracy_score(food_true + service_true + atmosphere_true, food_preds + service_preds + atmosphere_preds)
    overall_f1 = f1_score(food_true + service_true + atmosphere_true, food_preds + service_preds + atmosphere_preds, average="weighted")

    metrics = {
        "food_acc": food_acc, "food_f1": food_f1,
        "service_acc": service_acc, "service_f1": service_f1,
        "atmosphere_acc": atmosphere_acc, "atmosphere_f1": atmosphere_f1,
        "overall_acc": overall_acc, "overall_f1": overall_f1
    }

    return total_loss / len(data_loader), metrics
```

```python
from tabulate import tabulate

def display_test_results(test_loss, test_metrics):
    """
    Display test results (loss, accuracy, F1 score) in a tabular format.

    Args:
        test_loss (float): The loss value from the test evaluation.
        test_metrics (dict): A dictionary containing accuracy and F1 scores for each aspect.
                             Expected keys: 'food_acc', 'food_f1', 'service_acc', 'service_f1',
                                            'atmosphere_acc', 'atmosphere_f1'.
    """
    # Prepare data for the table
    table_data = [
        ["Aspect", "Accuracy", "F1 Score"],
        ["Food", f"{test_metrics['food_acc']:.4f}", f"{test_metrics['food_f1']:.4f}"],
        ["Service", f"{test_metrics['service_acc']:.4f}", f"{test_metrics['service_f1']:.4f}"],
        ["Atmosphere", f"{test_metrics['atmosphere_acc']:.4f}", f"{test_metrics['atmosphere_f1']:.4f}"],
        ["Overall", f"{test_metrics['overall_acc']:.4f}", f"{test_metrics['overall_f1']:.4f}"]
    ]

    # Print the table
    print(f"\nTest Loss: {test_loss:.4f}\n")

    table_str = tabulate(table_data, headers="firstrow", tablefmt="grid")
    print(table_str)

    return table_str
```

```python
import numpy as np
import math

batch_size = 64
learning_rate = 1e-5
epochs = 20

# Calculate the 95th Percentile Length
def calculate_p95_length(data, tokenizer):
    lengths = []
    for text in data["clean_cmt"]:
        tokenized = tokenizer(text, truncation=False, padding=False)
        lengths.append(len(tokenized["input_ids"]))

    p95_length = int(np.percentile(lengths, 95))
    # Calculate the next power of 2
    upper_power_of_2 = 2 ** math.ceil(math.log2(p95_length))

    return upper_power_of_2

# Calculate max_length for the dataset
max_length = calculate_p95_length(df, tokenizer)
print(f"95th percentile max_length: {max_length}")

# Use the calculated max_length in your dataset and dataloaders
train_dataset = ReviewDataset(train_data, tokenizer, max_length)
val_dataset = ReviewDataset(val_data, tokenizer, max_length)
test_dataset = ReviewDataset(test_data, tokenizer, max_length)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import os

def save_model(model, model_path):
    if not os.path.exists(model_path):
        os.makedirs(model_path)

    # Save model state_dict
    model_file_path = os.path.join(model_path, "model.pt")
    torch.save(model.state_dict(), model_file_path)
    print(f"Model saved to {model_file_path}")

    return model_file_path
```

```python
import torch.optim as optim
import torch.nn as nn
import torch

criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=learning_rate)

# Initialize the best metric
best_f1 = float('-inf')  # A very low initial value
best_loss = float('inf')

# Training loop
for epoch in range(epochs):
    # Train for one epoch
    train_loss = train_epoch(epoch, epochs, model, train_loader, optimizer, criterion, device)

    # Evaluate on validation set
    val_loss, val_metrics = evaluate_epoch(model, val_loader, criterion, device)
    display_test_results(val_loss, val_metrics)

    # Check if this is the best metric so far
    if val_metrics['overall_f1'] >= best_f1:
        if val_metrics['overall_f1'] == best_f1 and val_loss > best_loss:
            continue  # Skip if the loss is worse

        best_loss = val_loss  # Update the best loss
        best_f1 = val_metrics['overall_f1']  # Update the best metric

        save_model_path = save_model(model, absa_main_path)
        print(f"Epoch {epoch+1}: New best model saved with mean F1: {best_f1:.4f} | loss: {best_loss}")
```

# Save

```python
# Testing

model = AspectBasedSentimentModel(model_name, pretrained = False)
model.load_state_dict(torch.load(save_model_path, weights_only=True))

model.to(device)
model.eval()

print("Testing Model")
test_loss, test_metrics = evaluate_epoch(model, test_loader, criterion, device)
test_table = display_test_results(test_loss, test_metrics)

# Save test results to a text file
with open(path.join(absa_main_path, "eval.txt"), "w") as file:
    file.write(f"\nTest Loss: {test_loss:.4f}\n")
    file.write(test_table + "\n")
```
